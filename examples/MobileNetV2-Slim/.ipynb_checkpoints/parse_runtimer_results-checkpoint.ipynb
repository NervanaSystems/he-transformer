```python
import numpy as np
import numpy as np
import re
import os
import matplotlib.pyplot as plt
```

```python
def parse_runtime_file(filename):
    
    op_times = {}
    
    op_regex = r'.*\t(.*)_.* took (.*)ms'
    with open(filename) as f:
        lines = [line.strip() for line in f.readlines()]
        for line in lines:
            if re.match(op_regex, line):
                match = re.match(op_regex, line)
                print(match.groups())
                op_name = match.groups()[0]
                op_time = float(match.groups()[1])/1000.
                if op_name not in op_times:
                    op_times[op_name] = [op_time]
                else:
                    op_times[op_name].append(op_time)
    
    return op_times
```

```python
op_times = parse_runtime_file('./output.txt')
```

```python
op_times_sum = {}
for op in op_times:
    op_times_sum[op] = sum(op_times[op])
    
print(op_times_sum)

plt.bar(range(len(op_times_sum)), list(op_times_sum.values()), align='center')
plt.xticks(range(len(op_times_sum)), list(op_times_sum.keys()), rotation='45')
plt.ylabel('Runtime (s)')
plt.show()
```
