```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lines_13 = np.genfromtxt('gemm_13_results.txt', delimiter=',', dtype=None)
lines_14_strange = np.genfromtxt('gemm_14_results_strange.txt', delimiter=',', dtype=None)
lines_14_new = np.genfromtxt('gemm_14_results.txt', delimiter=',', dtype=None)

N13_p00 = [line for line in lines_13 if line[1] == 0.0]
N13_p05 = [line for line in lines_13 if line[1] == 0.5]
N13_p08 = [line for line in lines_13 if line[1] == 0.8]

print('l',len(lines_14_new))

N14_p00_strange = [line for line in lines_14_strange if line[1] == 0.0]
N14_p00_new = [line for line in lines_14_new if line[1] == 0.0]
print('N14_p00_new',N14_p00_new)
print('N14_p00_strange',N14_p00_strange)

N14_p00 = N14_p00_new + N14_p00_strange[len(N14_p00_new):]

print('N14_p00', N14_p00)
N14_p05 = [line for line in lines_14 if line[1] == 0.5]
N14_p08 = [line for line in lines_14 if line[1] == 0.8]

print([x[2] for x in N14_p00])
```

```python
plt.figure(figsize=(7,4))

ns = [5, 10, 15, 20, 25, 30, 35, 40, 45]

plt.plot([pt[0] for pt in N14_p00], [pt[2] for pt in N14_p00], 'o--', color='r', label='$N=2^{14},\ p_1 =0.0$')
plt.plot([pt[0] for pt in N14_p05], [pt[2] for pt in N14_p05], 'o--', color='b', label='$N=2^{14},\ p_1 =0.5$')
plt.plot([pt[0] for pt in N14_p08], [pt[2] for pt in N14_p08], 'o--', color='g', label='$N=2^{14},\ p_1 =0.8$')

plt.plot([pt[0] for pt in N13_p00], [pt[2] for pt in N13_p00], 'o-', color='r', label='$N=2^{13},\ p_1 =0.0$')
plt.plot([pt[0] for pt in N13_p05], [pt[2] for pt in N13_p05], 'o-', color='b', label='$N=2^{13},\ p_1 =0.5$')
plt.plot([pt[0] for pt in N13_p08], [pt[2] for pt in N13_p08], 'o-', color='g', label='$N=2^{13},\ p_1 =0.8$')

plt.legend(loc=0)
plt.xlabel('Square matrix dimension')
plt.ylabel('Runtime (sec)')

plt.savefig('GEMM.png', dpi=300, bbox_inches='tight')
```
